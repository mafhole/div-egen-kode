```python
df = pd.DataFrame({'year': [int(2018), int(2019), int(2020), int(2018), int(2017), int(2016)], 'termin': [1, 2, 3, 4, 5, 6], 
                   'oms': [1000, 1500, 2000, 5642, 5000, 800], 'orgb': ['123', '456', '789', '456', '754', '523']})
df
```
